```python
sales_per_year = df.groupby('YEAR_ID')['SALES'].sum().reset_index()
```

```python
plt.bar(sales_per_year['YEAR_ID'], sales_per_year['SALES'], color='blue')
plt.title('Total Sales by Year')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.xticks(sales_per_year['YEAR_ID'])  # Ensure all years are shown on x-axis
plt.show()
```

```python
sales_per_quarter = df.groupby('QTR_ID')['SALES'].sum().reset_index()

# Create a line plot
plt.plot(sales_per_quarter['QTR_ID'], sales_per_quarter['SALES'], marker='o', color='blue')
plt.title('Total Sales by Quarter')
plt.xlabel('Quarter ID')
plt.ylabel('Total Sales')
plt.xticks(sales_per_quarter['QTR_ID'])  # Ensure all quarters are shown on x-axis
plt.grid()
plt.show()
```

```python
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])

# Group by month and sum SALES
sales_over_time = df.resample('M', on='ORDERDATE')['SALES'].sum().reset_index()

# Create a line plot
plt.figure(figsize=(12, 6))  # Set figure size for better clarity
plt.plot(sales_over_time['ORDERDATE'], sales_over_time['SALES'], marker='o', color='blue')
plt.title('Total Sales Over Time (Monthly)', fontsize=16)
plt.xlabel('Order Date', fontsize=14)
plt.ylabel('Total Sales', fontsize=14)

# Formatting the x-axis for cleaner appearance
plt.xticks(sales_over_time['ORDERDATE'], rotation=45)
plt.grid()
plt.tight_layout()  # Adjust layout to make room for rotated labels
plt.show()
```

```python
sales_by_region = df.groupby('COUNTRY')['SALES'].sum().reset_index()

# Create a bar plot
plt.figure(figsize=(12, 6))
plt.bar(sales_by_region['COUNTRY'], sales_by_region['SALES'], color='skyblue')
plt.title('Sales Distribution by Country', fontsize=16)
plt.xlabel('Country', fontsize=14)
plt.ylabel('Total Sales', fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Group sales by country and product category
sales_by_region_category = df.groupby(['COUNTRY', 'PRODUCTLINE'])['SALES'].sum().unstack().fillna(0)

# Display the grouped data
print(sales_by_region_category)
```

```python
# Create a grouped bar plot
plt.figure(figsize=(12, 6))
sales_by_region_category.plot(kind='bar', ax=plt.gca(), color=plt.cm.tab20.colors)
plt.title('Total Sales by Country and Product Category', fontsize=16)
plt.xlabel('Country', fontsize=14)
plt.ylabel('Total Sales', fontsize=14)
plt.xticks(rotation=45)
plt.legend(title='Product Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
